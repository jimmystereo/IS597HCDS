```python
df_group = df.groupby(['link', 'model'])['score'].mean().reset_index()
df_group
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df_group is the DataFrame containing 'link', 'model', and 'score' (variance)
# Pivot the data to get one row per link and one column per model
df_pivot = df_group.pivot_table(index='link', columns='model', values='score')


# Calculate the pairwise absolute differences for each model pair
df_pivot['gpt_vs_claude'] = abs(df_pivot['gpt'] - df_pivot['claude'])
df_pivot['gpt_vs_gemini'] = abs(df_pivot['gpt'] - df_pivot['gemini'])
df_pivot['claude_vs_gemini'] = abs(df_pivot['claude'] - df_pivot['gemini'])

# Create subplots (3 subgraphs)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot the absolute difference distribution: GPT vs Claude
sns.histplot(df_pivot['gpt_vs_claude'], kde=True, ax=axes[0], color='blue', bins=30)
axes[0].set_title('Distribution of Absolute Variance Difference: GPT vs. Claude')
axes[0].set_xlabel('Absolute Variance Difference')
axes[0].set_ylabel('Frequency')

# Plot the absolute difference distribution: GPT vs Gemini
sns.histplot(df_pivot['gpt_vs_gemini'], kde=True, ax=axes[1], color='green', bins=30)
axes[1].set_title('Distribution of Absolute Variance Difference: GPT vs. Gemini')
axes[1].set_xlabel('Absolute Variance Difference')
axes[1].set_ylabel('Frequency')

# Plot the absolute difference distribution: Claude vs Gemini
sns.histplot(df_pivot['claude_vs_gemini'], kde=True, ax=axes[2], color='red', bins=30)
axes[2].set_title('Distribution of Absolute Variance Difference: Claude vs. Gemini')
axes[2].set_xlabel('Absolute Variance Difference')
axes[2].set_ylabel('Frequency')

# Adjust layout for better spacing
plt.tight_layout()

# Set same y-axis limit for all subplots
max_y = max(axes[0].get_ylim()[1], axes[1].get_ylim()[1], axes[2].get_ylim()[1])
for ax in axes:
    ax.set_ylim(0, max_y)

plt.show()
```

```python
import pandas as pd
from scipy import stats

# Pivot the df_group to have scores for each model in separate columns for each 'link'
df_pivot = df_group.pivot(index='link', columns='model', values='score')

# Ensure the pivoted DataFrame has the models as columns
models = ['claude', 'gemini', 'gpt']

# Iterate over each pair of models
for i in range(len(models)):
    for j in range(i+1, len(models)):
        model1 = models[i]
        model2 = models[j]

        # Get the scores for the two models
        model1_scores = df_pivot[model1]
        model2_scores = df_pivot[model2]

        # Paired t-test
        t_stat, t_p_value = stats.ttest_rel(model1_scores, model2_scores)

        # Pearson correlation coefficient
        correlation_coefficient, cor_p_value = stats.pearsonr(model1_scores, model2_scores)

        # Output the results
        print(f"--- Comparison: {model1} vs {model2} ---")
        print(f"Paired t-test: t-statistic = {t_stat}, p-value = {t_p_value}")
        print(f"Pearson Correlation Coefficient: {correlation_coefficient}, p-value = {cor_p_value}")

        # Interpret the results
        alpha = 0.05
        if t_p_value < alpha:
            print(f"Paired t-test: The difference between {model1} and {model2} is statistically significant.")
        else:
            print(f"Paired t-test: The difference between {model1} and {model2} is not statistically significant.")

        if cor_p_value < alpha:
            print(f"Pearson Correlation: The correlation between {model1} and {model2} is statistically significant.")
        else:
            print(f"Pearson Correlation: The correlation between {model1} and {model2} is not statistically significant.")

        print('===')
```

```python
grouped_pivot = df_group.pivot(index='link', columns='model', values='score')
# Create subplots
models = grouped_pivot.columns
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4), sharex=True, sharey=True)

colors = ['orange', 'skyblue', 'gray']  # Colors for the models
for idx, model in enumerate(models):
    variances = grouped_pivot[model].dropna()  # Exclude NaN values
    axes[idx].hist(variances, bins=20, color=colors[idx], alpha=1, edgecolor='black')
    if model == 'gpt':
        model = 'GPT'
    else:
        model = model.capitalize()
    axes[idx].set_title(f'Distribution of Score - {model}')
    axes[idx].grid(axis='y', alpha=0.75)
    axes[idx].set_ylabel('Frequency')
    axes[idx].set_xlabel('Score')
# Shared labels for the whole figure
# plt.xlabel('Variance')
plt.tight_layout()
plt.savefig('fig/h2.png')
plt.show()
```
